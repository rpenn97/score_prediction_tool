```python
# Overall DF

eng_df = pd.read_excel(r'EPL_understat.xlsx',sheet_name='gamesheet')
fra_df = pd.read_excel(r'Ligue1_understat.xlsx', sheet_name='gamesheet')
ger_df = pd.read_excel(r'Bundesliga_understat.xlsx', sheet_name='gamesheet')
ita_df = pd.read_excel(r'SerieA_understat.xlsx', sheet_name='gamesheet')
spa_df = pd.read_excel(r'LaLiga_understat.xlsx', sheet_name='gamesheet')

# Last 10 games Tables

eng_l10_table_h = pd.read_excel(r'EPL_understat.xlsx',sheet_name='last10_home')
eng_l10_table_a = pd.read_excel(r'EPL_understat.xlsx',sheet_name='last10_away')

fra_l10_table_h = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='last10_home')
fra_l10_table_a = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='last10_away')

ger_l10_table_h = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='last10_home')
ger_l10_table_a = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='last10_away')

ita_l10_table_h = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='last10_home')
ita_l10_table_a = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='last10_away')

spa_l10_table_h = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='last10_home')
spa_l10_table_a = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='last10_away')

# Home v Away Tables

eng_table_h = pd.read_excel(r'EPL_understat.xlsx',sheet_name='home')
eng_table_a = pd.read_excel(r'EPL_understat.xlsx',sheet_name='away')

fra_table_h = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='home')
fra_table_a = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='away')

ger_table_h = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='home')
ger_table_a = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='away')

ita_table_h = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='home')
ita_table_a = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='away')

spa_table_h = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='home')
spa_table_a = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='away')

# Ranked Tables
eng_r1_table_h = pd.read_excel(r'EPL_understat.xlsx',sheet_name='rank1_home')
eng_r2_table_h = pd.read_excel(r'EPL_understat.xlsx',sheet_name='rank2_home')
eng_r3_table_h = pd.read_excel(r'EPL_understat.xlsx',sheet_name='rank3_home')
eng_r1_table_a = pd.read_excel(r'EPL_understat.xlsx',sheet_name='rank1_away')
eng_r2_table_a = pd.read_excel(r'EPL_understat.xlsx',sheet_name='rank2_away')
eng_r3_table_a = pd.read_excel(r'EPL_understat.xlsx',sheet_name='rank3_away')

fra_r1_table_h = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='rank1_home')
fra_r2_table_h = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='rank2_home')
fra_r3_table_h = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='rank3_home')
fra_r1_table_a = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='rank1_away')
fra_r2_table_a = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='rank2_away')
fra_r3_table_a = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='rank3_away')

ger_r1_table_h = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='rank1_home')
ger_r2_table_h = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='rank2_home')
ger_r3_table_h = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='rank3_home')
ger_r1_table_a = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='rank1_away')
ger_r2_table_a = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='rank2_away')
ger_r3_table_a = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='rank3_away')

ita_r1_table_h = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='rank1_home')
ita_r2_table_h = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='rank2_home')
ita_r3_table_h = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='rank3_home')
ita_r1_table_a = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='rank1_away')
ita_r2_table_a = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='rank2_away')
ita_r3_table_a = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='rank3_away')

spa_r1_table_h = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='rank1_home')
spa_r2_table_h = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='rank2_home')
spa_r3_table_h = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='rank3_home')
spa_r1_table_a = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='rank1_away')
spa_r2_table_a = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='rank2_away')
spa_r3_table_a = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='rank3_away')

list_df = [eng_df,fra_df,ger_df,ita_df,spa_df]
ov_df = pd.concat(list_df)
ov_df['win'] = ov_df.apply(lambda x: 1 if x['Pts'] == 3 else 0, axis=1) 
ov_df['draw'] = ov_df.apply(lambda x: 1 if x['Pts'] == 1 else 0, axis=1) 
ov_df['lose'] = ov_df.apply(lambda x: 1 if x['Pts'] == 0 else 0, axis=1) 
ov_df['home?'] = ov_df.apply(lambda x: 1 if x['H/A'] == "h" else 0, axis=1)
ov_df

```

```python
def overall_table(df):
    # new df
    df_table = df.groupby("Team").agg({'played':'sum', 'xG':'sum', 'xGC':'sum','G':'sum', 'GC':'sum', 'Pts':'sum',
                                             'S':'sum', 'SC':'sum', 'QS': 'sum','QSC':'sum',
                                             'BTTS':'sum', 'Blank':'sum', 'CS':'sum','first_goal':'sum','first_conc':'sum'})
    df_table = df_table.apply(pd.to_numeric)
    df_table['GD'] = df_table['G'] - df_table['GC']
    df_table['xGD'] = df_table['xG'] - df_table['xGC']
    df_table['PPG'] = round(df_table['Pts'] / df_table['played'],2)
    df_table['BTTS_d'] = round(df_table['BTTS'] / df_table['played'],2)
    df_table['Blank_d'] = round(df_table['Blank'] / df_table['played'],2)
    df_table['CS_d'] = round(df_table['CS'] / df_table['played'],2)
    df_table['FG_d'] = round(df_table['first_goal'] / df_table['played'],2)
    df_table['FC_d'] = round(df_table['first_conc'] / df_table['played'],2)
    df_table = df_table[~(df_table['played'] <= 20)]
    df_table = df_table[['played','xG','xGC','xGD','G','GC','GD','Pts','PPG','S','SC','QS','QSC','BTTS_d','Blank_d','CS_d','FG_d','FC_d']]
    df_table = df_table.sort_values('PPG', ascending=False)
    df_table.insert(0,'Placement',range(1, 1+ len(df_table)))

    return df_table

eng_ov_table = overall_table(eng_df)
fra_ov_table = overall_table(fra_df)
ger_ov_table = overall_table(ger_df)
ita_ov_table = overall_table(ita_df)
spa_ov_table = overall_table(spa_df)

league_list = [eng_ov_table, fra_ov_table, ger_ov_table, ita_ov_table, spa_ov_table]
ov_table = pd.concat(league_list)
ov_table
```

```python
# Goal Differential vs PPG
y = ov_table.PPG
x = ov_table.GD.values.reshape(-1,1)
model = LinearRegression().fit(x,y)
r_sq = model.score(x,y)
intercept = model.intercept_
slope = model.coef_
y_pred = intercept + (slope*x)
fig, ax = plt.subplots(figsize=(10,10))
plt.scatter(x,y)
plt.ylim(0.4,2.55)
plt.plot(x,y_pred,c='red')
plt.yticks([0.5,1,1.5,2,2.5])
# plt.gca().invert_yaxis()
plt.xlabel('GD')
plt.ylabel('Points')
plt.title(f'LR on relationship between Goal Difference and Points Per Game ({r_sq}x)')
```

```python
def combine_and_net(home_df,away_df):
    df = pd.DataFrame({'Home PPG' : home_df['PPG'], 'Away PPG': away_df['PPG']})
    for team in df:
        df['net'] = df['Home PPG'] - df['Away PPG']
    df2 = df.drop(['Home PPG', 'Away PPG'], axis=1)
    df2.loc['AVERAGE'] = df2.mean()
    df = df.drop('net', axis=1)
    df.loc['AVERAGE'] = df.mean()

    return df, df2
```
