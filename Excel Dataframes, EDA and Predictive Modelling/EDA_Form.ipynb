```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
eng_df = pd.read_excel(r'EPL_understat.xlsx',sheet_name='gamesheet')
eng_df = eng_df.sort_values(by='date',ascending=False)
fra_df = pd.read_excel(r'Ligue1_understat.xlsx',sheet_name='gamesheet')
fra_df = fra_df.sort_values(by='date',ascending=False)
ger_df = pd.read_excel(r'Bundesliga_understat.xlsx',sheet_name='gamesheet')
ger_df = ger_df.sort_values(by='date',ascending=False)
ita_df = pd.read_excel(r'SerieA_understat.xlsx',sheet_name='gamesheet')
ita_df = ita_df.sort_values(by='date',ascending=False)
spa_df = pd.read_excel(r'LaLiga_understat.xlsx',sheet_name='gamesheet')
spa_df = spa_df.sort_values(by='date',ascending=False)
```

```python
teamlist_eng = []
teamlist_fra = []
teamlist_ger = []
teamlist_ita = []
teamlist_spa = []
```

```python
def teamlist(df,teamlist):
    team_array = df["Team"].unique()
    for team in team_array:
        teamlist.append(team)
    teamlist = sorted(set(teamlist))
    return teamlist
```

```python
teamlist_eng = teamlist(eng_df, teamlist_eng)
teamlist_fra = teamlist(fra_df, teamlist_fra)
teamlist_ger = teamlist(ger_df, teamlist_ger)
teamlist_ita = teamlist(ita_df, teamlist_ita)
teamlist_spa = teamlist(spa_df, teamlist_spa)
# teamlist_spa
```

```python
def last10(df):

    temp_df = df
    temp_df['date'] = pd.to_datetime(temp_df['date'])
    temp_df['date_int'] = (datetime.datetime.now()) - (temp_df['date'])
    temp_df = temp_df.groupby(['Team'],as_index=False).apply(lambda x: x.nsmallest(10, 'date_int'))
    temp_df.reset_index(inplace=True)
    temp_df.drop(['level_0','level_1','date_int'],axis=1,inplace=True)
    return temp_df
```

```python
eng_l10 = last10(eng_df)
with pd.ExcelWriter("../Excel Dataframes, EDA and Predictive Modelling/EPL_last10.xlsx") as writer:
    eng_l10.to_excel(writer, sheet_name="last10")
eng_l10
```
